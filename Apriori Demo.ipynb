```python
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Crypto.csv',header=None)
data.isnull().sum()
```

```python
transaction=[]
for i in range(data.shape[0]):
    transaction.append([str(data.values[i,j])for j in range(data.shape[1])])
transaction=np.array(transaction)
transaction
```

```python
te=TransactionEncoder()
te_ary=te.fit(transaction).transform(transaction)
dataset=pd.DataFrame(te_ary,columns=te.columns_)
dataset
```

```python
dataset.shape
```

```python
dataset.columns
```

```python
def encode_units(x):
    if x==False:
        return 0;
    if x==True:
        return 1;
dataset=dataset.applymap(encode_units)
dataset.head()
```

```python
frequent_itemset=apriori(dataset,min_support=0.05,use_colnames=True)

frequent_itemset['length']=frequent_itemset['itemsets'].apply(lambda x:len(x))
frequent_itemset.head()
```

```python
frequent_itemset.tail()
```

```python
frequent_itemset[(frequent_itemset['length']==2)].head()
```

```python
rules=association_rules(frequent_itemset,metric='lift',min_threshold=1.4)
rules
```

```python
rules['antecedents_length']=rules['antecedents'].apply(lambda x:len(x))
rules['consequents_length']=rules['consequents'].apply(lambda x:len(x))
rules.sort_values('lift',ascending=False)
```

```python
rules.drop(rules.columns[[7,8]],axis=1,inplace=True)
```
